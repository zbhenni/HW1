I split each question into three different coding windows.

Question 1: Lower energies will be more affected by the potential. They will either bounce off the reflective core or they could become a bound state. Higher energies will be less affected by the potential. Very high energies can even pass close to the potential with very minimal deflection.

```python
# Initialization Code used in all the programs. This code was taken from the classes and modified a little.

%matplotlib inline
import numpy as np
from matplotlib import pyplot
from matplotlib.colors import ColorConverter as cc
import math

class particle2(object):
    
    def __init__(self, mass=1., x=0., y=0., vx=0., vy=0., fx = 0., fy = 0.):
        self.mass = mass
        self.x = x
        self.y = y
        self.vx = vx
        self.vy = vy
        self.fx = fx     # Added the self.fx and self.fy
        self.fy = fy
        
    def get_force(self, fx, fy, V_0, a):  # returns force per unit of mass (acceleration)
        r = math.sqrt(self.x*self.x+self.y*self.y)
        a_r7 = pow(a/r,7.)/a                 # the 'a' value is the location of the minimum and can be set in the main code
        a_r13 = pow(a/r,13.)/a
        self.fx = 4*V_0*self.x*(12.*a_r13-6.*a_r7)   # Lenard-Jones potenials
        self.fy = 4*V_0*self.y*(12.*a_r13-6.*a_r7)
        self.fx /= self.mass
        self.fy /= self.mass
        
    def verlet(self, dt, V_0, a):
        self.get_force(self.fx, self.fy, V_0, a) # before I move to the new position
        self.x += self.vx*dt + 0.5*self.fx*dt*dt
        self.y += self.vy*dt + 0.5*self.fy*dt*dt
        self.vx += 0.5*self.fx*dt
        self.vy += 0.5*self.fy*dt
        self.get_force(self.fx, self.fy, V_0, a) # after I move to the new position
        self.vx += 0.5*self.fx*dt
        self.vy += 0.5*self.fy*dt
```

Question 2: Below the deflection function is plotted against b values. The initial energy (E_i) was specified in the beginning.

It was specified: E_i = $0.1*V_0$

```python
# Question 2: b vs. scattering angle (fixed E_i)

m1 = 0.1
V_0 = 1.                    # Amount of potential: input as a factor of V_0

E_i = V_0              # Initial energy in terms of the potential


v1 = math.sqrt(2.*E_i/m1)    # Initial Velocity
r1 = -20.                     # initial r value
b = 1.                       # initial b value
a = 1.                       # set the location of the minimum in the Lenard-Jones potential 

dt = 0.001                    # time step
tmax = 30.
nsteps = int(tmax/dt)

# Create Arrays to store position and velocity values
x = np.zeros(shape=(nsteps,1)) # r distance away
y = np.zeros(shape=(nsteps,1)) # b value
vx = np.zeros(shape=(nsteps,1)) 
vy = np.zeros(shape=(nsteps,1))

# Initial position and velocity Values except position y
x[0,0] = r1
vx[0,0] = v1
vy[0,0] = 0.

# Scan over different b values
b_max = 5.
db = .1
bsteps = int(b_max/db)
angle_b = np.zeros(shape=(bsteps,1))


# n will loop over different b's
for n in range(0,bsteps):
    
    # Initial y position specified by b. Note: it has (1+n) b/c when n=0, y[0,0]=db
    y[0,0] = (1+n)*db
    
    
    # specify object class 'particle2'
    p_in = [particle2]
    p_in.append([particle2])

    # Set initial values in object class; calculate initial force
    m = [m1]
    p_in[0] = particle2(m[0], x[0,0], y[0,0], vx[0,0], vy[0,0])    

    r = math.sqrt(x[0,0]*x[0,0]+y[0,0]*y[0,0])
    a_r7 = pow(a/r,7.)/a                 # the 'a' value is the location of the minimum and can be set in the main code
    a_r13 = pow(a/r,13.)/a
    fx = 4*V_0*x[0,0]*(12.*a_r13-6.*a_r7)   # Lenard-Jones potenials
    fy = 4*V_0*y[0,0]*(12.*a_r13-6.*a_r7)

    # Calculate path of particle at current b and E_i value
    for i in range(1,nsteps):
        p_in[0].verlet(dt, V_0, a)

        x[i,0] = p_in[0].x
        y[i,0] = p_in[0].y
        vx[i,0] = p_in[0].vx
        vy[i,0] = p_in[0].vy        
    
    if x[-1,0]>0:
        angle_b[n] = math.atan(y[-1,0]/x[-1,0])*180./math.pi
    else:
        if y[-1,0]>0:
            angle_b[n] = math.atan(y[-1,0]/x[-1,0])*180./math.pi+180.
        else: 
            angle_b[n] = math.atan(y[-1,0]/x[-1,0])*180./math.pi-180.


b_array = np.linspace(db,b_max,bsteps)


if x[-1,0]>0:
    angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi
else:
    if y[-1,0]>0:
        angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi+180.
    else: 
        angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi-180.
        
print 'The final angle is: {} degrees with b= {}'.format(angle, b_max-db)

pyplot.plot(b_array[:], angle_b[:], color='blue', ls='-', lw=3)
pyplot.xlabel('b value (units)')
pyplot.ylabel('angle (degrees)');
```

```python
# Question 3: E vs. Scattering Angle (Fixed b)

V_0 = 1.                    # Amount of potential: input as a factor of V_0
m1 = 0.1
r1 = -10.                     # initial r value

b = .1                       # initial b value
a = 1.                       # set the location of the minimum in the Lenard-Jones potential 

dt = 0.001                    # time step
tmax = 20.
nsteps = int(tmax/dt)

# Create Arrays to store position and velocity values
x = np.zeros(shape=(nsteps,1)) # r distance away
y = np.zeros(shape=(nsteps,1)) # b value
vx = np.zeros(shape=(nsteps,1)) 
vy = np.zeros(shape=(nsteps,1))

# Initial position and velocity Values except position y
x[0,0] = r1
y[0,0] = b
vy[0,0] = 0.

# Scan over different E_i values
Ei_max = 50.*V_0
dEi = .1*V_0
Eisteps = int(Ei_max/dEi)
angle_b = np.zeros(shape=(Eisteps,1))


# n will loop over different E_i's
for n in range(0,Eisteps):
    
    # Initial velocity specified by E_i. Note: it has (1+n) b/c when n=0, E_i = .1*V_0
    vx[0,0] = math.sqrt(2.*(1+n)*dEi/m1)
    
    # specify object class 'particle2'
    p_in = [particle2]
    p_in.append([particle2])

    # Set initial values in object class; calculate initial force
    m = [m1]
    p_in[0] = particle2(m[0], x[0,0], y[0,0], vx[0,0], vy[0,0])    

    r = math.sqrt(x[0,0]*x[0,0]+y[0,0]*y[0,0])
    a_r7 = pow(a/r,7.)/a                 # the 'a' value is the location of the minimum and can be set in the main code
    a_r13 = pow(a/r,13.)/a
    fx = 4*V_0*x[0,0]*(a_r13-a_r7)   # Lenard-Jones potenials
    fy = 4*V_0*y[0,0]*(a_r13-a_r7)

    # Calculate path of particle at current b and E_i value
    for i in range(1,nsteps):
        p_in[0].verlet(dt, V_0, a)

        x[i,0] = p_in[0].x
        y[i,0] = p_in[0].y
        vx[i,0] = p_in[0].vx
        vy[i,0] = p_in[0].vy        
    
    if x[-1,0]>0:
        angle_b[n] = math.atan(y[-1,0]/x[-1,0])*180./math.pi
    else:
        if y[-1,0]>0:
            angle_b[n] = math.atan(y[-1,0]/x[-1,0])*180./math.pi+180.
        else: 
            angle_b[n] = math.atan(y[-1,0]/x[-1,0])*180./math.pi-180.

    
Ei_array = np.linspace(dEi,Ei_max,Eisteps)


if x[-1,0]>0:
    angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi
else:
    if y[-1,0]>0:
        angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi+180.
    else: 
        angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi-180.

print 'The final angle is: {} degrees with E_i= {}*V_0'.format(angle, Ei_max-dEi)

pyplot.plot(Ei_array[:], angle_b[:], color='blue', ls='-', lw=3)
pyplot.xlabel('Initial Energy Expressed as a Fraction of the Potential, V_0 (units)')
pyplot.ylabel('angle (degrees)');
```

Question 3 part b: Calculate the differential cross section vs. b. This was done using the following equation: $$\frac{d\sigma}{d\Omega}=\frac{b}{\sin{\Theta}}\left|\frac{db}{d\Theta}\right|.
$$The initial energy was kept fixed at E_i = $0.01*V_0$

```python
# Question 3: Differential Cross Section vs. b value (fixed initial energy)

V_0 = 1.                    # Amount of potential: input as a factor of V_0

m1 = 0.1
r1 = -20.                     # initial r value

E_i = .01*V_0              # Initial energy in terms of the potential
v1 = math.sqrt(2.*E_i/m1)    # Initial Velocity
a = 1.                       # set the location of the minimum in the Lenard-Jones potential 

dt = 0.001                    # time step
tmax = 5.
nsteps = int(tmax/dt)

# Create Arrays to store position and velocity values
x = np.zeros(shape=(nsteps,1)) # r distance away
y = np.zeros(shape=(nsteps,1)) # b value
vx = np.zeros(shape=(nsteps,1)) 
vy = np.zeros(shape=(nsteps,1))

# Initial position and velocity Values except position y
x[0,0] = r1
vx[0,0] = v1
vy[0,0] = 0.

# Scan over different b values
b_max = 10.
db = .1
bsteps = int(b_max/db)
angle_b = np.zeros(shape=(bsteps,1))


# n will loop over different b's
for n in range(0,bsteps):
    
    # Initial y position specified by b. Note: it has (1+n) b/c when n=0, y[0,0]=db
    y[0,0] = (1+n)*db
    
    
    # specify object class 'particle2'
    p_in = [particle2]
    p_in.append([particle2])

    # Set initial values in object class; calculate initial force
    m = [m1]
    p_in[0] = particle2(m[0], x[0,0], y[0,0], vx[0,0], vy[0,0])    

    r = math.sqrt(x[0,0]*x[0,0]+y[0,0]*y[0,0])
    a_r7 = pow(a/r,7.)/a                 # the 'a' value is the location of the minimum and can be set in the main code
    a_r13 = pow(a/r,13.)/a
    fx = 4*V_0*x[0,0]*(12.*a_r13-6.*a_r7)   # Lenard-Jones potenials
    fy = 4*V_0*y[0,0]*(12.*a_r13-6.*a_r7)

    # Calculate path of particle at current b and E_i value
    for i in range(1,nsteps):
        p_in[0].verlet(dt, V_0, a)

        x[i,0] = p_in[0].x
        y[i,0] = p_in[0].y
        vx[i,0] = p_in[0].vx
        vy[i,0] = p_in[0].vy        
    
    if x[-1,0]>0:
        if y[-1,0]>0:
            angle_b[n] = math.atan(y[-1,0]/x[-1,0])*180./math.pi
        else: 
            angle_b[n] = -math.atan(y[-1,0]/x[-1,0])*180./math.pi
    else:       
        if y[-1,0]>0:   
            angle_b[n] = math.atan(y[-1,0]/x[-1,0])*180./math.pi+180.
        else:
            angle_b[n] = -math.atan(y[-1,0]/x[-1,0])*180./math.pi+180.


b_array = np.linspace(db*2,b_max-db,bsteps-2)
scat_cross = np.zeros(shape=(bsteps-2,1))

# Calculate differential scattering cross section. This calculates the equation above
for i in range(0, bsteps-2):
    scat_cross[i] = math.fabs(((1+i)*db/math.sin(angle_b[i+1]*math.pi/180))*math.fabs(2.*db/((angle_b[i+2]-angle_b[i])*math.pi/180.)))
    
    
if x[-1,0]>0:
    angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi
else:
    if y[-1,0]>0:
        angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi+180.
    else: 
        angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi-180.
    
print 'The final angle is: {} degrees with b= {}'.format(angle, b_max-db)

pyplot.plot(b_array[:], scat_cross[:], color='blue', ls='-', lw=3)
pyplot.xlabel('b value (units)')
pyplot.ylabel('Differential Scattering Cross Section (units)');
```

Question 3 part c: Calculate the differential cross section vs. E_i. This was done using the following equation: $$\frac{d\sigma}{d\Omega}=\frac{b}{\sin{\Theta}}\left|\frac{db}{d\Theta}\right|.
$$The initial energy was kept fixed at E_i = $0.01*V_0$

```python
# Question 3: Differential Cross Section vs. initial energy (fixed b value)

V_0 = 1.                    # Amount of potential: input as a factor of V_0

m1 = 0.1
r1 = -20.                     # initial r value

E_i = .01*V_0              # Initial energy in terms of the potential
v1 = math.sqrt(2.*E_i/m1)    # Initial Velocity
a = 1.                       # set the location of the minimum in the Lenard-Jones potential 

dt = 0.01                    # time step
tmax = 16.
nsteps = int(tmax/dt)

# Create Arrays to store position and velocity values
x = np.zeros(shape=(nsteps,1)) # r distance away
y = np.zeros(shape=(nsteps,1)) # b value
vx = np.zeros(shape=(nsteps,1)) 
vy = np.zeros(shape=(nsteps,1))

# Initial position and velocity Values except position y
x[0,0] = r1
vx[0,0] = v1
vy[0,0] = 0.

# Scan over 3 different b values, so the differential deflection function can be found in terms of b
b = .2
db = .01
b_max = b+db
b_min = b-db
bsteps = 3
angle_b = np.zeros(shape=(bsteps,1))


# Scan over different E_i values
Ei_max = 10.*V_0
dEi = .1*V_0
Eisteps = int(Ei_max/dEi)
scat_cross = np.zeros(shape=(Eisteps,1))


#------ Calculate the differential scattering cross section vs. energy START-------#


for j in range(0,Eisteps):
    # Initial velocity specified by E_i. Note: it has (1+n) b/c when n=0, E_i = .1*V_0
    vx[0,0] = math.sqrt(2.*(1+j)*dEi/m1)

    # n will loop over different b's
    for n in range(0,bsteps):
    
        # Initial y position specified by b. Note: it has (1+n) b/c when n=0, y[0,0]=db
        y[0,0] = b_min+db*n
    
        # specify object class 'particle2'
        p_in = [particle2]
        p_in.append([particle2])

        # Set initial values in object class; calculate initial force
        m = [m1]
        p_in[0] = particle2(m[0], x[0,0], y[0,0], vx[0,0], vy[0,0])    

        r = math.sqrt(x[0,0]*x[0,0]+y[0,0]*y[0,0])
        a_r7 = pow(a/r,7.)/a                 # the 'a' value is the location of the minimum and can be set in the main code
        a_r13 = pow(a/r,13.)/a
        fx = 4*V_0*x[0,0]*(12.*a_r13-6.*a_r7)   # Lenard-Jones potenials
        fy = 4*V_0*y[0,0]*(12.*a_r13-6.*a_r7)

        # Calculate path of particle at current b and E_i value
        for i in range(1,nsteps):
            p_in[0].verlet(dt, V_0, a)

            x[i,0] = p_in[0].x
            y[i,0] = p_in[0].y
            vx[i,0] = p_in[0].vx
            vy[i,0] = p_in[0].vy        
    
        if x[-1,0]>0:
            angle_b[n] = math.atan(y[-1,0]/x[-1,0])*180./math.pi
        else:
            if y[-1,0]>0:
                angle_b[n] = math.atan(y[-1,0]/x[-1,0])*180./math.pi+180.
            else: 
                angle_b[n] = math.atan(y[-1,0]/x[-1,0])*180./math.pi-180.


    # Calculate differential scattering cross section. This calculates the equation above
    scat_cross[j] = (b/math.sin(angle_b[1]*math.pi/180))*math.fabs((2.*db)/((angle_b[2]-angle_b[0])*180./math.pi))
    
#------ Calculate the differential scattering cross section vs. energy END-------#


Ei_array = np.linspace(dEi,Ei_max,Eisteps) # array for plotting
    
if x[-1,0]>0:
    angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi
else:
    if y[-1,0]>0:
        angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi+180.
    else: 
        angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi-180.
    
print 'The final angle is: {} degrees with b= {}'.format(angle, b_max-db)

pyplot.plot(Ei_array[:], scat_cross[:], color='blue', ls='-', lw=3)
pyplot.xlabel('Initial Energy Expressed as a Fraction of the Potential, V_0 (units)')
pyplot.ylabel('Differential Scattering Cross Section (units)');
```

Question 4: In this question we were asked to determine the maximum energy for which the Lennard-Jones potential exhibits orbiting by solving the correct set of equations involving $V$ and its derivatives.

We can see from the above graph that around E_i = $0.5*V_0$ the graph spikes upwards, when going from right to left. We then plot a = 1., b = .1, and E_i = $0.5*V_0$. This exhibited orbiting. I then manually changed the value of E_i and ran the code again. After several iterations I found the max E_i that exhibits orbiting is: E_i = $.53206*V_0$.

Of course this can be taken to more decimal places; however, then I would need to change my step size to see the new accuracy.

```python
# Simple Scattering

V_0 = 1.                    # Amount of potential: input as a factor of V_0

m1 = 0.1
r1 = -1.                   # initial r value
b = 0.1                    # initial b value

E_i = .53206*V_0               # Initial energy in terms of the potential
v1 = math.sqrt(2.*E_i/m1)   # Initial Velocity
a = 1.                      # set the location of the minimum in the Lenard-Jones potential 

dt = 0.01                   # time step
tmax = 3000.
nsteps = int(tmax/dt)

# Create Arrays to store position and velocity values
x = np.zeros(shape=(nsteps,1)) # r distance away
y = np.zeros(shape=(nsteps,1)) # b value
vx = np.zeros(shape=(nsteps,1)) 
vy = np.zeros(shape=(nsteps,1))

# Initial position and velocity Values except position y
x[0,0] = r1
vx[0,0] = v1
vy[0,0] = 0.

# Initial y position specified by b
y[0,0] = b

# specify object class 'particle2'
p_in = [particle2]
p_in.append([particle2])

# Set initial values in object class; calculate initial force
m = [m1]
p_in[0] = particle2(m[0], x[0,0], y[0,0], vx[0,0], vy[0,0])    

r = math.sqrt(x[0,0]*x[0,0]+y[0,0]*y[0,0])
a_r7 = pow(a/r,7.)/a                 # the 'a' value is the location of the minimum and can be set in the main code
a_r13 = pow(a/r,13.)/a
fx = 4*V_0*x[0,0]*(12.*a_r13-6.*a_r7)   # Lenard-Jones potenials
fy = 4*V_0*y[0,0]*(12.*a_r13-6.*a_r7)

# Calculate path of particle at current b and E_i value
for i in range(1,nsteps):
    p_in[0].verlet(dt, V_0, a)

    x[i,0] = p_in[0].x
    y[i,0] = p_in[0].y
    vx[i,0] = p_in[0].vx
    vy[i,0] = p_in[0].vy        


if x[-1,0]>0:
    angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi
else:
    if y[-1,0]>0:
        angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi+180.
    else: 
        angle = math.atan(y[-1,0]/x[-1,0])*180./math.pi-180.


print 'final y position: {}. Final x position: {}' .format(y[-1,0], x[-1,0])
print 'The final angle is: {} degrees with b= {}'.format(angle, y[0,0])
print 'It can be seen that the particle is still close, indicating it is bound.'

pyplot.plot(x[:,0], y[:,0], color='blue', ls='-', lw=3)
pyplot.xlabel('x position')
pyplot.ylabel('y position');
```
